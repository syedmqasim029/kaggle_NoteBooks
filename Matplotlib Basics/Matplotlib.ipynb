## Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4]
b = [10,20,30,40]
plt.plot(a,b)
plt.show()
```

```python
plt.bar(a,b)
plt.xlabel("Numbers")
plt.ylabel("Steps")
plt.show()
```

```python
x = ['python','cpp','java','c']
y = [87,75,82,66]
plt.bar(x,y,color=['r','g','b','m'],width=0.8,align='edge')
# align : to set each bar at edge or center
plt.title("Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

```python

x = ['python','cpp','java','c']
y = [87,75,82,66]
plt.bar(x,y,color='r',width=0.8,align='center',edgecolor='b',linewidth=6,linestyle=':')
plt.title("Programming Languages",fontsize=15)
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.show()
```

```python
x = ['python','cpp','java','c']
y = [87,75,82,66]
plt.bar(x,y,color='g',width=0.8,edgecolor='b',linewidth=4,alpha=0.7)
# alpha : to control visiblity of bar
plt.title("Programming Languages",fontsize=15)
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.show()
```

```python
x = ['python','cpp','java','c']
y = [87,75,82,66]
plt.bar(x,y,color=['g','r','r','g'],width=0.8,edgecolor='b',linewidth=4,alpha=0.7,label='Popular')
# alpha : to control visiblity of bar
plt.legend()
# legent to show label
plt.title("Programming Languages",fontsize=15)
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.show()
```

```python
import numpy as np
x = np.array(['python','cpp','java','c'])
y = np.array([87,75,82,66])
plt.bar(x[[0,2]],y[[0,2]],color='g',width=0.8,edgecolor='b',linewidth=4,alpha=0.7,label='Popular')
plt.bar(x[[1,3]],y[[1,3]],color='r',width=0.8,edgecolor='b',linewidth=4,alpha=0.7,label='Old')
plt.legend()
plt.title("Programming Languages",fontsize=15)
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.show()
# just used numpy for fancy indexing
```

```python
import numpy as np
x = np.array(['python','cpp','java','c'])
y = np.array([87,75,82,66])
z = np.array([30,70,55,85])
width = 0.4
p = np.arange(len(x))
p1=[i+width for i in p]
plt.bar(p,y,color='g',width=width,alpha=0.7,label='2000-2026')
plt.bar(p1,z,color='r',width=width,alpha=0.7,label='1980-2000')
plt.title("Programming Languages",fontsize=15)
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
a = np.array([53, 36, 46, 41, 12, 18, 27, 45, 33, 25, 22, 12,
              40, 20, 15, 36, 34, 30, 54, 42, 50, 22, 10, 46,
              20, 59, 53, 48, 31, 10, 45, 59, 26, 50, 14, 53,
              28, 56, 29, 42, 28, 20, 12, 47, 49, 33, 59, 48,
              33, 35])

# Create figure with specific size
plt.figure(figsize=(12, 8))

# Histogram with new parameters
n, bins, patches = plt.hist(
    a,
    bins=12,  # Explicitly set number of bins
    color='steelblue',
    edgecolor='darkblue',
    linewidth=2,
    alpha=0.7,  # Transparency
    density=True,  # Normalize to form probability density
    histtype='stepfilled',  # Different histogram style
    label='Data Distribution',
    hatch='/'  # Pattern fill
)

# Highlight specific bins (e.g., highlight bins with values > 15)
for i, patch in enumerate(patches):
    if bins[i] > 45:  # Highlight high value bins
        patch.set_facecolor('coral')
        patch.set_alpha(0.9)
        patch.set_edgecolor('darkred')
        patch.set_linewidth(3)

# Title and labels with new formatting
plt.title('ENHANCED HISTOGRAM ANALYSIS',
          fontsize=18,
          fontweight='bold',
          color='darkgreen',
          pad=20)

plt.xlabel('Data Values', fontsize=14, fontweight='semibold')
plt.ylabel('Probability Density', fontsize=14, fontweight='semibold')

# Add multiple vertical lines with different styles
plt.axvline(np.mean(a), color='red', linestyle='--', linewidth=3,
            label=f'Mean: {np.mean(a):.1f}')
plt.axvline(np.median(a), color='green', linestyle='-.', linewidth=3,
            label=f'Median: {np.median(a):.1f}')
plt.axvline(35, color='blue', linestyle='-', linewidth=2,
            label='Reference (35)')

# Add percentiles
percentile_25 = np.percentile(a, 25)
percentile_75 = np.percentile(a, 75)
plt.axvline(percentile_25, color='orange', linestyle=':', linewidth=2,
            alpha=0.7, label=f'25th %ile: {percentile_25:.1f}')
plt.axvline(percentile_75, color='purple', linestyle=':', linewidth=2,
            alpha=0.7, label=f'75th %ile: {percentile_75:.1f}')

# Add fill between percentiles (IQR)
plt.axvspan(percentile_25, percentile_75, alpha=0.1, color='gray',
            label='IQR Region')

# Add text annotations
plt.text(np.mean(a) + 2, max(n) * 0.9, f' = {np.std(a):.1f}',
         fontsize=12, bbox=dict(boxstyle="round,pad=0.3",
                                facecolor="yellow", alpha=0.7))

# Add a horizontal line at 0
plt.axhline(y=0, color='black', linewidth=0.5, alpha=0.5)

# Customize grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.grid(True, axis='y', alpha=0.3)

# Legend with customization
plt.legend(loc='upper right', fontsize=10, framealpha=0.9,
           shadow=True, fancybox=True)

# Add secondary x-axis showing data range
plt.twiny()  # Create a second x-axis on top
plt.xlim(plt.xlim())  # Match limits
plt.xticks([])  # Remove ticks
plt.xlabel(f'Data Range: {np.min(a)} - {np.max(a)} | n = {len(a)}',
           fontsize=10, style='italic')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Add a footnote
plt.figtext(0.5, 0.01,
            f'Statistics: Min={np.min(a)}, Max={np.max(a)}, '
            f'Range={np.max(a)-np.min(a)}',
            ha='center', fontsize=9, style='italic',
            bbox=dict(boxstyle="round,pad=0.2", facecolor="lightgray", alpha=0.5))

plt.show()

#Just Leaning new things from this graph
```

## Pie Chart

```python
x = [10,20,30,40]
y = ['C','C++','Java','Python']
c = ['r', 'g', 'b', 'y']
plt.pie(x,labels=y,colors=c,autopct="%0.1f%%")
plt.title('Languages',fontweight='bold',fontsize=15)
plt.show()
```

```python
x = [10,20,30,40]
y = ['C','C++','Java','Python']
c = ['red', 'blue', 'green', 'khaki']
plt.pie(x,labels=y,colors=c,autopct="%0.1f%%",shadow=True,explode=[0.2,0,0,0],radius=1.1)
plt.title('Languages',fontweight='bold',fontsize=15)
plt.show()
```

```python
x = [10,20,30,40]
y = ['C','C++','Java','Python']
c = ['red', 'blue', 'green', 'khaki']
plt.pie(x,labels=y,colors=c,autopct="%0.1f%%",shadow=True,explode=[0.2,0,0,0],radius=1.1,labeldistance=1.1,textprops={'fontsize':12})
plt.title('Programming Languages Popularity',fontweight='bold',fontsize=15)
plt.show()
```

```python
x = [10, 20, 30, 40]
labels = ['C', 'C++', 'Java', 'Python']
colors = ['red', 'blue', 'green', 'khaki']
explode = [0.2, 0.05, 0.05, 0.05]

plt.figure(figsize=(8, 8))

plt.pie(x,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',pctdistance=0.75,labeldistance=1.1,shadow=True,startangle=90,
radius=1.1,counterclock=False,wedgeprops={'edgecolor': 'black','linewidth': 2},textprops={'fontsize': 12,'color': 'black','fontweight': 'bold'})

plt.title('Programming Languages Popularity',fontsize=16,fontweight='bold',color='purple',pad=20)

plt.legend(labels,title='Languages',loc='center left',bbox_to_anchor=(1, 0.5),fontsize=10,title_fontsize=12)

plt.axis('equal')

plt.show()
```
