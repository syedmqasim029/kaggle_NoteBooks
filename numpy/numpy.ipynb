```python
# Deep ans shallow copy
a = np.arange(1,16)
slice = a[2:10]
slice = slice * 10
```

```python
slice # shallow copy
```

```python
a # no change in the array
```

```python
b = a
```

```python
b[5] = 99 # deep copy
```

```python
a # since the elelment 5 is changed now 
```

```python
b 
```

```python
b = a.copy() # to avoid use copy function
b[5] = 1002
```

```python
a
```

```python
# Matrix multiplication
a = np.array([[1,2,3],[4,5,6]]) #2,3
```

```python
b = np.array([[7,8,9],[10,11,12],[5,6,7]]) # 3,3
```

```python
a @ b # first mat col should be = row of other mat here 3=3
```

```python
np.dot(a,b) # same as a @ b
```

```python
# stacking 
np.vstack((a,b)) # vertical stack
```

```python
b = np.array([[7,8,9],[10,11,12],[5,6,7]]) 
a = np.array([[7,8,9],[10,11,12],[5,6,7]])
A = np.hstack((a,b))
A
```
